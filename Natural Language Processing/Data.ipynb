```python
import requests
from bs4 import BeautifulSoup
import pickle

def url_to_textbit(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    text = [p.text for p in soup.find(class_="article-body").find_all('p')]
    print(url)
    return text

def url_to_textbit_bbc(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    text = [p.text for p in soup.find(class_="story-body__inner").find_all('p')]
    print(url)
    return text

def url_to_textbit_theguardian(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    text = [p.text for p in soup.find(class_="content__article-body from-content-api js-article__body").find_all('p')]
    print(url)
    return text

def url_to_textbit_thestar(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    text = [p.text for p in soup.find(class_="c-article-body__content").find_all('p')]
    print(url)
    return text

def url_to_textbit_dailymail(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    text = [p.text for p in soup.find(id="js-article-text").find_all('p')]
    print(url)
    return text

def url_to_textbit_latimes(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    text = [p.text for p in soup.find(class_="RichTextArticleBody-body RichTextBody").find_all('p')]
    print(url)
    return text

def url_to_textbit_newsau(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    text = [p.text for p in soup.find(class_="story-content").find_all('p')]
    print(url)
    return text

urls_fox=['https://www.foxnews.com/us/boston-rally-against-massachusetts-coronavirus-restriction',
'https://www.foxnews.com/media/tucker-carlson-beijing-coronavirus-new-chinese-century',
'https://www.foxnews.com/us/coronavirus-sparks-unprecedented-struggle-for-nycs-homeless-shelters',
'https://www.foxnews.com/politics/ohio-mike-dewine-obnoxious-protesters',
'https://www.foxnews.com/us/nypd-officers-say-coronavirus-social-distancing-policing-needs-to-end']

urls_bbc=['https://www.bbc.com/news/health-52475688',
'https://www.bbc.com/news/uk-52540068?intlink_from_url=https://www.bbc.com/news/coronavirus&link_location=live-reporting-story',
'https://www.bbc.com/news/health-52354520',
'https://www.bbc.com/news/world-middle-east-52493608',
'https://www.bbc.com/news/science-environment-52529830']

urls_theguardian=['https://www.theguardian.com/us-news/2020/apr/04/how-long-does-coronavirus-live-on-different-surfaces',
'https://www.theguardian.com/world/2020/apr/01/what-is-coronavirus-and-what-is-the-mortality-rate-covid-19',
'https://www.theguardian.com/world/2020/feb/21/who-is-most-at-risk-of-contracting-coronavirus']
#,'https://www.theguardian.com/world/2020/may/04/mexico-president-amlo-coronavirus-briefings-like-trump'
#need one more article

urls_thestar=['https://www.thestar.com/news/investigations/2020/05/05/they-said-there-was-no-playbook-for-dealing-with-covid-19-outbreak-at-nursing-homes-there-were-several.html',
'https://www.thestar.com/politics/federal/2020/05/02/could-the-blood-of-recovered-covid-19-patients-be-used-to-treat-the-virus-canada-is-launching-a-trial-to-find-out.html',
'https://www.thestar.com/business/2020/04/17/covid-19-disruption-has-some-parents-worried-theyll-lose-out-on-maternity-benefits.html',
'https://www.thestar.com/politics/federal/2020/04/09/did-you-get-covid-19-money-from-ottawa-this-week-heres-who-you-can-thank.html',
'https://www.thestar.com/news/canada/2020/04/22/what-does-reaching-the-peak-in-the-covid-19-pandemic-mean.html']

urls_dailymail=['https://www.dailymail.co.uk/tvshowbiz/article-8286459/Julia-Roberts-stuns-tiered-tulle-black-white-gown-honor-Met-Gala.html',
'https://www.dailymail.co.uk/news/article-8287433/NHS-Covid-contact-tracing-app-goes-live-Isle-Wight-today.html',
'https://www.dailymail.co.uk/news/article-8287761/Unions-threaten-stage-mass-walkouts-lockdown-easing-risks.html',
'https://www.dailymail.co.uk/news/article-8285985/Piers-Morgan-gets-CLEAR-coronavirus-stay-work.html',
'https://www.dailymail.co.uk/news/article-8287615/Iranian-airline-fuelled-Middle-Easts-coronavirus-crisis-continuing-flights-China.html']

urls_latimes=['https://www.latimes.com/science/newsletter/2020-05-01/newsom-pressure-stay-home-orders-coronavirus-today',
'https://www.latimes.com/science/newsletter/2020-04-29/isolation-mental-health-coronavirus-today',
'https://www.latimes.com/science/newsletter/2020-04-24/heat-wave-beaches-coronavirus-today',
'https://www.latimes.com/science/newsletter/2020-04-22/second-wave-infections-protests-california-coronavirus-today',
'https://www.latimes.com/science/newsletter/2020-04-20/health-workers-la-antibody-test-coronavirus-today']

urls_nytimes =['https://www.nytimes.com/2020/05/05/world/coronavirus-news.html#link-77471a29',
'https://www.nytimes.com/2020/05/05/world/coronavirus-news.html#link-264fe9ac',
'https://www.nytimes.com/2020/05/04/business/coronavirus-china-vaccine.html',
'https://www.nytimes.com/2020/05/03/business/car-maintenance-tips-coronavirus-lockdown.html',
'https://www.nytimes.com/2020/05/05/arts/insurance-claims-coronavirus-arts.html']

urls_newsau=['https://www.news.com.au/world/coronavirus/global/coronavirus-united-states-and-china-race-to-find-covid19-vaccine/news-story/ada519c9e81ae49a68f2025359e853d9#itm=newscomau%7Cnews%7Cright-now-in-%7C0%7CCoronavirus%3A%20United%20States%20and%20China%20race%20to%20find%20COVID-19%20v...%7Cindex%7Cworld-coronavirus&itmt=1588669995924',
'https://www.news.com.au/world/coronavirus/australia/people-who-refuse-to-download-the-covidsafe-virus-tracing-app-are-the-new-antivaxxers/news-story/541c36fe5cdb56eb1a098b0b9a0dddcc#itm=newscomau%7Cnews%7Cright-now-in-%7C1%7CCOVIDSafe%3A%20No%20excuse%20to%20not%20download%20virus%20tracing%20app%7Cindex%7Cworld-coronavirus&itmt=1588669998225',
'https://www.news.com.au/world/coronavirus/global/president-trumps-theory-on-origin-of-coronavirus-dismissed-by-who-health-experts/news-story/07fff449547220197be2084f7d9033b0#itm=newscomau%7Cnews%7Cright-now-in-%7C2%7CPresident%20Trump%E2%80%99s%20theory%20on%20origin%20of%20coronavirus%20dismissed%20...%7Cindex%7Cworld-coronavirus&itmt=1588670000013',
'https://www.news.com.au/world/coronavirus/closures/coronavirus-australia-northern-territory-is-the-best-state-complying-with-restrictions/news-story/455f6ff74392ef5dc472785c3faae510#itm=newscomau%7Cnews%7Cright-now-in-%7C3%7CCoronavirus%20Australia%3A%20Northern%20Territory%20is%20the%20best%20state%20...%7Cindex%7Cworld-coronavirus&itmt=1588670000509',
'https://www.news.com.au/world/coronavirus/australia/people-who-refuse-to-download-the-covidsafe-virus-tracing-app-are-the-new-antivaxxers/news-story/541c36fe5cdb56eb1a098b0b9a0dddcc']

urls = ['https://www.foxnews.com/us/boston-rally-against-massachusetts-coronavirus-restriction',
'https://www.foxnews.com/media/tucker-carlson-beijing-coronavirus-new-chinese-century',
'https://www.foxnews.com/us/coronavirus-sparks-unprecedented-struggle-for-nycs-homeless-shelters',
'https://www.foxnews.com/politics/ohio-mike-dewine-obnoxious-protesters',
'https://www.foxnews.com/us/nypd-officers-say-coronavirus-social-distancing-policing-needs-to-end',
'https://www.bbc.com/news/health-52475688',
'https://www.bbc.com/news/uk-52540068?intlink_from_url=https://www.bbc.com/news/coronavirus&link_location=live-reporting-story',
'https://www.bbc.com/news/health-52354520',
'https://www.bbc.com/news/world-middle-east-52493608',
'https://www.bbc.com/news/science-environment-52529830',
'https://www.theguardian.com/us-news/2020/apr/04/how-long-does-coronavirus-live-on-different-surfaces',
'https://www.theguardian.com/world/2020/apr/01/what-is-coronavirus-and-what-is-the-mortality-rate-covid-19',
'https://www.theguardian.com/world/2020/feb/21/who-is-most-at-risk-of-contracting-coronavirus',
'https://www.theguardian.com/world/2020/may/04/mexico-president-amlo-coronavirus-briefings-like-trump',
'https://www.theguardian.com/business/live/2020/may/05/uk-car-sales-tumble-services-pmi-recession-covid-19-business-live',
'https://www.thestar.com/news/investigations/2020/05/05/they-said-there-was-no-playbook-for-dealing-with-covid-19-outbreak-at-nursing-homes-there-were-several.html',
'https://www.thestar.com/politics/federal/2020/05/02/could-the-blood-of-recovered-covid-19-patients-be-used-to-treat-the-virus-canada-is-launching-a-trial-to-find-out.html',
'https://www.thestar.com/business/2020/04/17/covid-19-disruption-has-some-parents-worried-theyll-lose-out-on-maternity-benefits.html',
'https://www.thestar.com/politics/federal/2020/04/09/did-you-get-covid-19-money-from-ottawa-this-week-heres-who-you-can-thank.html',
'https://www.thestar.com/news/canada/2020/04/22/what-does-reaching-the-peak-in-the-covid-19-pandemic-mean.html',
'https://www.dailymail.co.uk/tvshowbiz/article-8286459/Julia-Roberts-stuns-tiered-tulle-black-white-gown-honor-Met-Gala.html',
'https://www.dailymail.co.uk/news/article-8287433/NHS-Covid-contact-tracing-app-goes-live-Isle-Wight-today.html',
'https://www.dailymail.co.uk/news/article-8287761/Unions-threaten-stage-mass-walkouts-lockdown-easing-risks.html',
'https://www.dailymail.co.uk/news/article-8285985/Piers-Morgan-gets-CLEAR-coronavirus-stay-work.html',
'https://www.dailymail.co.uk/news/article-8287615/Iranian-airline-fuelled-Middle-Easts-coronavirus-crisis-continuing-flights-China.html',
'https://www.latimes.com/science/newsletter/2020-05-01/newsom-pressure-stay-home-orders-coronavirus-today',
'https://www.latimes.com/science/newsletter/2020-04-29/isolation-mental-health-coronavirus-today',
'https://www.latimes.com/science/newsletter/2020-04-24/heat-wave-beaches-coronavirus-today',
'https://www.latimes.com/science/newsletter/2020-04-22/second-wave-infections-protests-california-coronavirus-today',
'https://www.latimes.com/science/newsletter/2020-04-20/health-workers-la-antibody-test-coronavirus-today',
'https://www.news.com.au/world/coronavirus/global/coronavirus-united-states-and-china-race-to-find-covid19-vaccine/news-story/ada519c9e81ae49a68f2025359e853d9#itm=newscomau%7Cnews%7Cright-now-in-%7C0%7CCoronavirus%3A%20United%20States%20and%20China%20race%20to%20find%20COVID-19%20v...%7Cindex%7Cworld-coronavirus&itmt=1588669995924',
'https://www.news.com.au/world/coronavirus/australia/people-who-refuse-to-download-the-covidsafe-virus-tracing-app-are-the-new-antivaxxers/news-story/541c36fe5cdb56eb1a098b0b9a0dddcc#itm=newscomau%7Cnews%7Cright-now-in-%7C1%7CCOVIDSafe%3A%20No%20excuse%20to%20not%20download%20virus%20tracing%20app%7Cindex%7Cworld-coronavirus&itmt=1588669998225',
'https://www.news.com.au/world/coronavirus/global/president-trumps-theory-on-origin-of-coronavirus-dismissed-by-who-health-experts/news-story/07fff449547220197be2084f7d9033b0#itm=newscomau%7Cnews%7Cright-now-in-%7C2%7CPresident%20Trump%E2%80%99s%20theory%20on%20origin%20of%20coronavirus%20dismissed%20...%7Cindex%7Cworld-coronavirus&itmt=1588670000013',
'https://www.news.com.au/world/coronavirus/closures/coronavirus-australia-northern-territory-is-the-best-state-complying-with-restrictions/news-story/455f6ff74392ef5dc472785c3faae510#itm=newscomau%7Cnews%7Cright-now-in-%7C3%7CCoronavirus%20Australia%3A%20Northern%20Territory%20is%20the%20best%20state%20...%7Cindex%7Cworld-coronavirus&itmt=1588670000509'
'https://www.news.com.au/world/coronavirus/australia/people-who-refuse-to-download-the-covidsafe-virus-tracing-app-are-the-new-antivaxxers/news-story/541c36fe5cdb56eb1a098b0b9a0dddcc'
]

newspaper = ['foxnews', 'bbc', 'theguardian', 'thestar', 'dailymail', 'latimes', 'news.au']
```

```python
def combine_text(list_of_text):
    combined_text =  ' '.join(list_of_text)
    return combined_text
```

```python
data_combined = {key:[combine_text(value)] for (key,value) in data.items()}
```

```python
import pandas as pd
pd.set_option('max_colwidth', 150)

data_df = pd.DataFrame.from_dict(data_combined).transpose()
data_df.columns = ['textbits']
data_df = data_df.sort_index()
data_df
```

```python
data_df.textbits.loc['dailymail']
```

```python
#start up some cleaning
#læs op på lamdba expressions

import re
import string

def clean_text_round1(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w', '', text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
data_clean = pd.DataFrame(data_df.textbits.apply(round1))
data_clean
```

```python
data_clean.textbits.loc['bbc']
```

```python
#more cleaning, after checking what is still missing
def clean_text_round2(text):
    text = re.sub('[´`–]', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('[’“”—]', '', text)
    #text = re.sub('...', ' ', text)
    #text = re.sub('ei', '', text)
    return text

round2 = lambda x: clean_text_round2(x)
```

```python
data_clean = pd.DataFrame(data_clean.textbits.apply(round2))
data_clean
```
