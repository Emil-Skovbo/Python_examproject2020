## Most common words

```python
import pandas as pd

data = pd.read_pickle('dtm.pkl')
data = data.transpose()
data.head()
```

```python
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(30)
    top_dict[c] = list(zip(top.index, top.values))

top_dict
```

```python
#For better view
for newspaper, top_words in top_dict.items():
    print(newspaper)
    print(', '.join([word for word, count in top_words[0:14]]))
    print('---')
```

```python
#we pull out most common words so we cann ad them to our stopwords
words = []
for newspaper in data.columns:
    top = [word for (word,count) in top_dict[newspaper]]
    for t in top:
        words.append(t)        
words
```

```python
from collections import Counter

Counter(words).most_common()
```

```python
#Adding more stop words

add_stop_words = [word for word, count in Counter(words).most_common() if count > 2]
add_stop_words
```

```python
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer
import pickle

data_clean = pd.read_pickle('data_clean.pkl')
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(data_clean.textbits)
data_stop = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_stop.index=data_clean.index

#Update file
#pickle.dump(cv, open('cv_stop_.pkl', 'wb'))
#data_stop.to_pickle('dtm_stop_.pkl')
```

## WordClouds 

```python
from wordcloud import WordCloud

wc = WordCloud(stopwords=stop_words, background_color="white", colormap="Dark2", max_font_size=150, random_state=42)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [16,6]

newspapernames = ['fox','bbc','theguardian','thestar','dailymail','newsau']

for index, newspapernames in enumerate(data.columns):
    wc.generate(data_clean.textbits[newspapernames])
    
    plt.subplot(3,4, index+1)
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(newspapernames)

plt.show()
    
```
